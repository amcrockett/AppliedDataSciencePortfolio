```python
#remove columns not needed
yelp.drop(['address', 'attributes', 'hours', 'is_open', 'latitude', 'longitude', 'neighborhood'], axis=1, inplace=True)
```

```python
yelp.head()
```

```python
#select only businesses that are restaurants
yelp_cleaned = yelp[yelp['categories'].str.contains('Restaurants') == True]
```

```python
yelp_cleaned.head()
```

```python
#select only rows that contain a 5 digit postal code
yelp_cleaned = yelp_cleaned[yelp_cleaned['postal_code'].str.contains('^(\d{5})$') == True]
```

```python
yelp_cleaned.head()
```

```python
#transform categories out of multi value
test_cat = yelp_cleaned.set_index('business_id').categories.str.split(', ', expand=True).stack()
```

```python
#reset index
test_cat = test_cat.reset_index()
```

```python
test_cat.head()
```

```python
#inspect categories remaining
test_cat.groupby(0)['business_id'].count()
```

```python
restaurants_only = test_cat.groupby(0)['business_id'].count()
```

```python
#inspect categories only where value > 13
restaurants_only[restaurants_only > 13]
```

```python
restaurants_only = restaurants_only.reset_index()
```

```python
print(len(restaurants_only[0]), 'categories remain')
```

```python
#rename columns
test_cat = test_cat.rename(index=str, columns={0: "category"})
test_cat = test_cat.rename(index=str, columns={'level_1': "instance"})
```

```python
test_cat.head()
```

```python
businesses_final = pd.merge(yelp_cleaned, test_cat, on='business_id', how = 'inner')
```
